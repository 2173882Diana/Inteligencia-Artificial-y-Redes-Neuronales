```python
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dropout, Dense
import tensorflow as tf
```

```python
dataset = pd.read_csv('water_potability dataset 1.csv')
dataset.head()
```

```python
print(dataset.isnull().sum())
```

```python
dataset['ph'] = dataset['ph'].fillna(dataset['ph'].mean())
```

```python
X = dataset[["ph", "Conductivity", "Turbidity"]]
y = dataset["Potability"]
```

```python
Potability_names = ["No potable", "Potable"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
sX = StandardScaler()
X_train = sX.fit_transform(X_train)
X_test = sX.transform(X_test)
```

```python
   from tensorflow.keras.callbacks import EarlyStopping

clf=Sequential()
clf.add(Dense(units=2,kernel_initializer='uniform',activation='relu',input_dim=3))
clf.add(Dropout(rate=0.1))
clf.add(Dense(units=2,kernel_initializer='uniform',activation='relu'))
clf.add(Dropout(rate=0.1))
clf.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
clf.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
# The initial fit call without early stopping
#clf.fit(X_train,y_train,batch_size=10,epochs=100)

# Initialize EarlyStopping
early_stopping = EarlyStopping(monitor='accuracy', patience=10, restore_best_weights=True)

# Fit the model again with early stopping, using the test set for validation
clf.fit(
        X_train, y_train,
        epochs=100,
        validation_data=(X_test, y_test),
        callbacks=[early_stopping]
    )
```
